# Chapter 2 — Probability Theory on Coin Toss Space

## Overview
This chapter develops the probability-theoretic foundations needed for 
rigorous asset pricing. Key concepts are introduced in the finite binomial 
model and will be extended to continuous time in later chapters.

## Key Concepts

### 1. Probability Space
A finite probability space $(\Omega, \mathcal{F}, \mathbb{P})$ satisfying 
the Kolmogorov axioms (Ex2.1). The sample space $\Omega$ consists of all 
possible coin toss sequences $\omega_1\omega_2\cdots\omega_N$.

### 2. Conditional Expectation
$\mathbb{E}_n[X]$ is itself a random variable — its value depends on the 
first $n$ coin tosses. This is the discrete-time analogue of 
$\mathbb{E}[X|Y]$ in classical statistics.

### 3. Martingale
A process $M_0, M_1, \ldots, M_N$ satisfying:
$$\mathbb{E}_n[M_{n+1}] = M_n$$
**Key results:**
- Discounted stock price $\frac{S_n}{(1+r)^n}$ is a martingale under $\tilde{\mathbb{P}}$ (Theorem 2.4.4)
- Discounted wealth process $\frac{X_n}{(1+r)^n}$ is a martingale under $\tilde{\mathbb{P}}$ (Theorem 2.4.5)
- A convex function of a martingale is a submartingale (Ex2.3)
- A discrete-time stochastic integral of a martingale is a martingale (Ex2.6)

### 4. Markov Process
A process where the current state is sufficient to determine future behavior:
$$\mathbb{E}_n[f(X_{n+1})] = g(X_n)$$
**Key results:**
- Stock price $S_n$ is Markov (Theorem 2.5.7)
- Lookback option price is martingale but NOT Markov (Ex2.8)
- $(S_n, Y_n)$ is Markov for path-dependent options (Ex2.13, 2.14)

### 5. Risk-Neutral Pricing
The time-$n$ price of a derivative with payoff $V_N$ is:
$$V_n = \tilde{\mathbb{E}}_n\left[\frac{V_N}{(1+r)^{N-n}}\right]$$
**Key results:**
- Equivalent to Chapter 1 backward induction (Ex2.8)
- Holds for dividend-paying stocks (Ex2.10)
- Cash flow valuation (Theorem 2.4.8)

### 6. Important Tools
- **Tower Property:** $\mathbb{E}_n[\mathbb{E}_m[X]] = \mathbb{E}_n[X]$ for $n \leq m$
- **Independence Lemma:** Past and future are independent given $\mathcal{F}_n$
- **Jensen's Inequality:** $\mathbb{E}[\varphi(X)] \geq \varphi(\mathbb{E}[X])$ for convex $\varphi$

# Exercise 2.1 - Properties of Probabilit

### Background
**Definition 2.1.1 (Kolmogorov Axioms): ** A probability measure $\mathbb{P}$ on a finite space $\Omega$ satisfies:
1. $\mathbb{P}(\Omega) = 1$
2. $\mathbb{P}(A) \geq 0$ for all all events $A$
3. If $A \cap B = \emptyset$, then $\mathbb{P}(A \cup B) = \mathbb{P}(A) + \mathbb{P}(B)$

### Problem
Using Definition 2.1.1 show the following:

**(i)** If $A$ is an event and $A^c$ denotes its complement, then $\mathbb{P}(A^c) = 1 - \mathbb{P}(A)$.

**(ii)** If $A_1, A_2, \ldots, A_N$ is a finite set of events, then
$$\mathbb{P}\left(\bigcup_{n=1}^{N} A_n\right) \leq \sum_{n=1}^{N} \mathbb{P}(A_n) \tag{2.8.1}$$
If the events $A_1, A_2, \ldots, A_N$ are disjoint, then equality holds in (2.8.1).

### Proof
**(i)** Since $A \cup A^c = \Omega$ and $A \cap A^c = \emptyset$, by Axiom(3):
$$\mathbb{P}(A \cup A^c) = \mathbb{P}(A) + \mathbb{P}(A^c)$$
By Axiom(1), $\mathbb{P}(\Omega) = 1$, therefore:
$$\mathbb{P}(A^c) = 1 - \mathbb{P}(A) \quad \blacksquare$$

**(ii)** **Base case** $N=2$: Since $A_1 \cup A_2 = A_1 \cup (A_2 \setminus A_1)$ and these are disjoint:
$$\mathbb{P}(A_1 \cup A_2) = \mathbb{P}(A_1) + \mathbb{P}(A_2 \setminus A_1)$$
Since $A_2 \setminus A_1 \subseteq A_2$, by Axiom(2):
$$\mathbb{P}(A_2 \setminus A_1) \leq \mathbb{P}(A_2)$$
Therefore:
$$\mathbb{P}(A_1 \cup A_2) \leq \mathbb{P}(A_1) + \mathbb{P}(A_2)$$

**Inductive step**: Assume the inequality holds for $N-1$ events. Then:
$$\mathbb{P}\left(\bigcup_{n=1}^{N} A_n\right) = \mathbb{P}\left(\bigcup_{n=1}^{N-1} A_n \cup A_n\right) \leq \mathbb{P}\left(\bigcup_{n=1}^{N-1} A_n\right) + \mathbb{P}(A_N)$$
$$\leq \sum_{n=1}^{N-1} \mathbb{P}(A_n) + \mathbb{P}(A_N) = \sum_{n=1}^{N} \mathbb{P}(A_n) \quad \blacksquare$$

**Equality (disjoint case)**: If $A_i \cap A_j = \emptyset$ for all $i \neq j$, then $\mathbb{P}(A_i \cap A_j) = 0$, so  $\mathbb{P}(A_2 \setminus A_1) = \mathbb{P}(A_2)$ and equality holds throughout. $\blacksquare$

# Exercise 2.2 - Distribution and Expectation of Stcok Price

### Setop
From Figure 2.3.1: $S_0 = 4, u = 2, d = \frac{1}{2}, r = \frac{1}{4}$

### (i) Distribution of $S_3$ under $\tilde{\mathbb{P}}$ ($\tilde{p} = \tilde{q} = \frac{1}{2}$)

| $S_3$ | Probability |
|-------|-------------|
| $32$  |$\frac{1}{8}$|
|  $8$  |$\frac{3}{8}$|
|  $2$  |$\frac{3}{8}$|
|$0.5$  |$\frac{1}{8}$|

### (ii) Expectation under $\tilde{\mathbb{P}}$

$$\tilde{\mathbb{E}}[S_1] = 5, \quad \tilde{\mathbb{E}}[S_2] = 6.25, \quad \tilde{\mathbb{E}}[S_3] = \frac{125}{16}$$

Average growth rate per period:
$$\frac{\tilde{\mathbb{E}}[S_1]}{S_0} = \frac{\tilde{\mathbb{E}}[S_2]}{\tilde{\mathbb{E}}[S_1]} = \frac{\tilde{\mathbb{E}}[S_3]}{\tilde{\mathbb{E}}[S_2]} = \frac{5}{4} = 1 + r$$

**under $\tilde{\mathbb{P}}$, the average growth rate equals the risk-free rate $r$.**
This is the defining property of the risk-neutral measure.

### (iii) Distribution of $S_3$ under $\mathbb{P}$ ($p = \frac{2}{3}, q = \frac{1}{3}$)

| $S_3$ | Probability |
|------|-------------|
| $32$ |$\frac{8}{27}$|
|  $8$ |$\frac{4}{9}$ |
|  $2$ |$\frac{2}{9}$ |
|$0.5$ |$\frac{1}{27}$|

$$\mathbb{E}[S_1] = 6, \quad \mathbb{E}[S_2] = 9, \quad \mathbb{E}[S_3] = \frac{27}{2}$$

Average growth rate per period:
$$\frac{\mathbb{E}[S_1]}{S_0} = \frac{\mathbb{E}[S_2]}{\mathbb{E}[S_1]} = \frac{\mathbb{E}[S_3]}{\mathbb{E}[S_2]} = \frac{3}{2} = pu + qd$$

**Under $\mathbb{P}$, the average growth rate equals $pu + qd > 1 + r$.**
Stock earns a risk premium above the risk-free rate under the real-world measure.

### Key Insight

| Measure | Average Growth Rate | Interpretation|
|---------|---------------------|---------------|
|$\tilde{\mathbb{P}}$ (risk-neutral)| $1 + r = \frac{5}{4}$ |Risk-free rate|
|$\mathbb{P}$ (real-world) | $pu + qd = \frac{3}{2}$ | Risk premium above $r$|

# Exercise 2.3 - Convex Function of a Martingale is a Submartingale

### Problem 
Let $M_0. M_1, \ldots, M_N$ be a martingale and let $\varphi$ be a convec function.
Show that $\varphi(M_0), \varphi(M_1), \ldots, \varphi(M_N)$ is a submartingale.

### Definitions
**Martingale:** $\mathbb{E}_n[M_{n+1}] = M_n$

**Submartingale:** $\mathbb{E}_n[\varphi(M_{n+1})] \geq \varphi(M_n)$

### Proof
By Jensen's Inequality (Theorem 2.2.5) and the Martingale property:
 $$\mathbb{E}_n[\varphi(M_{n+1})] \geq \varphi(\mathbb{E}_n[M_{n+1}]) = \varphi(M_n) \quad \blacksquare$$

### Examples
Common convec functions applied to martingales:

| $\varphi(x)$ | Result|
|-------------|--------|
| $x^2$ | $M_n^2$ is a submartingale|
| $e^x$ | $e^{M_n}$ is a submartingale|
| $\max(x-K, 0)$ | Discounted option price is a submartingale under $\tilde{\mathbb{P}}$|

# Exercise 2.4 - Symmetric Random Walk as a Martingale

### Setup
Let $X_j = +1$ (head) or $-1$ (tail) with probability $\frac{1}{2}$ each.
Define $M_0 = 0$ and $M_n = \sum_{j=1}^{n}X_j$ (symmetric random walk).

### (i) $M_n$ is a Martingale

Since $M_{n+1} = M_n + X_{n+1}$ and $X_{n+1}$ is independent of $M_n$:

$$\mathbb{E}_n[M_{n+1}] = \frac{1}{2}(M_n + 1) + \frac{1}{2}(M_n - 1) = M_n \quad \blacksquare$$

### (ii) $S_n$ is a Martingale

Define:
$$S_n = e^{\sigma M_n} \left(\frac{2}{e^{\sigma} + e^{-\sigma}}\right)^n$$

Using $M_{n+1} = M_n + X_{n+1}$ and independence of $X_{n+1}$:

$$\mathbb{E}_n[S_{n+1}] = e^{\sigma M_n} \left(\frac{2}{e^{\sigma}+e^{-\sigma}}\right)^n \cdot \frac{2}{e^{\sigma}+e^{-\sigma}} \cdot \mathbb{E}_n[e^{\sigma X_{n+1}}]$$

Since $\mathbb{E}_n[e^{\sigma X_{n+1}}] = \frac{1}{2}e^{\sigma} + \frac{1}{2}e^{-\sigma} = \frac{e^\sigma + e^{-\sigma}}{2}$:

$$\mathbb{E}_n[S_{n+1}] = S_n \cdot \frac{2}{e^\sigma+e^{-\sigma}}\cdot\frac{e^\sigma + e^{-\sigma}}{2} = S_n \quad \blacksquare$$

### Key Insight
The discount factor $\frac{2}{e^\sigma+s^{-\sigma}}$ is exactly the reciprocal of  $\mathbb{E}[e^{\sigma X}]$, which corrects for the upward drift introduced by the convex exponential function(cf. Exercise 2.3).

# Exercise 2.5 - Itô's Formula for Random Walk and Markov Property of $I_n$

### Setup
Let $M_n$ be the symmetric random walk of Exercise 2.4.
Define $I_0 = 0$ and:
$$I_n = \sum_{j=0}^{n-1} M_j(M_{j+1} - M_j), \quad n = 1, 2, \ldots$$

### (i) $I_n = \frac{1}{2}M_n^2 - \frac{n}{2}$

Using the identity:
$$M_{j+1}^2 - M_j^2 = (M_{j+1} - M_j)^2 + 2M_j(M_{j+1} - M_j) = 1+ 2M_j(M_{j+1} - M_j)$$

Therefore:
$$2I_n = \sum_{j=0}^{n-1}(M_{j+1}^2 - M_j^2) - \sum_{j=0}^{n-1}1$$

The first sum telescopes:
$$\sum_{j=0}^{n-1}(M_{j+1}^2 - M_j^2) = M_n^2 - M_0^2 = M_n^2$$

Hence:
$$I_n = \frac{1}{2}M_n^2 - \frac{n}{2} \quad \blacksquare$$

### (ii) $I_n$ is a Markov Process

Using part(i): $I_{n+1} = \frac{1}{2}M_{n+1}^2 - \frac{n+1}{2}$

Substituting $M_{n+1} = M_n + X_{n+1}$:

$$\mathbb{E}_n[f(I_{n+1})] = \frac{1}{2}f\left(\frac{1}{2}(M_n+1)^2 - \frac{n+1}{2}\right) + \frac{1}{2}f\left(\frac{1}{2}(M_n-1)^2 - \frac{n+1}{2}\right)$$

$$= \frac{1}{2}f\left(I_n + M_n\right) + \frac{1}{2}f\left(I_n - M_n\right)$$

Since $M_n^2 = 2I_n + n$, we have $|M_n| = \sqrt{2I_n + n}$.
The two terms are symmetric in $\pm M_n$, so $M_n$ drops out:

$$\mathbb{E}_n[f(I_{n+1})] = g(I_n)$$

where:

$$g(i) = \frac{1}{2}f\left(i + \sqrt{2i+n}\right) + \frac{1}{2}\left(i - \sqrt{2i+n}\right) \quad \blacksquare$$

Since $\mathbb{E}_n[f(I_{n+1})]$ depends only on $I_n$ (and the constant $n$),
$I_0, I_1, I_2, \ldots$ is a Markov process.

### Key Insight
This is the discrete-time analogue of **Itô's formula**:
$$I_n = \frac{1}{2}M_n^2 - \frac{n}{2} \longleftrightarrow \int_0^t W_s\, dW_s = \frac{1}{2}W_t^2 - \frac{t}{2}$$

where $W_t$ is Browninan mothin. This connection becomes central in Chapter 4.

# Exercise 2.6 - Discrete-Time Stochastic Integral is a Martingale

### Setup
Let $M_0, M_1, \ldots, M_N$ be a martingale and $\Delta_0, \Delta_1, \ldots, \Delta_{N-1}$ be an adapted process. Define $I_0 = 0$ and:

$$I_n = \sum_{j=0}^{n-1} \Delta_j(M_{j+1} - M_j), \quad n = 1, \ldots, N$$

### Problem
Show that $I_0, I_1, \ldots, I_n$ is a martingale.

### Proof 
Since $I_{n+1} - I_{n} = \Delta_n(M_{n+1} - M_n)$:

$$\mathbb{E}_n[I_{n+1}] = \mathbb{E}_n[I_n + \Delta_n(M_{n+1} - M_n)]$$

$$= I_n + \Delta_n(\mathbb{E}_n[M_{n+1} - M_n)$$

$$= I_n + \Delta_n \cdot 0 = I_n \quad \blacksquare$$

where:
- $I_n$ is $\mathcal{F}_n$-measurable -> taken outside $\mathbb{E}_n$
- $\Delta_n$ is adapted -> known at time $n$, taken outside $\mathbb{E}_n$
- $\mathbb{E}_n[M_{n+1}] = M_n$ -> Martingale property of $M_n$

### Key Insight
This is the discrete-time analogue of the continuous-time result:

$$\int_0^t \Delta_s dW_s \text{ is a martingale}$$

which is fundamental to stochastic calculus in Chapter 4.
The adapted condition on $\Delta_n$ is essential - it ensures we are not "looking into the fure# when choosing our position.

# Exercicse 2.8 - A Martingale that is Not Markov

### Example: Lookback Option

In the binomial model with $S_0=4, u=2, d=\frac{1}{2}, r=\frac{1}{4}$, consider the Lookback Call Option with payoff:

$$V_3 = \max_{0 \leq n \leq 3} S_n - S_3$$

**$V_n$ is a Martingale**

By the risk-neutral pricing formula, the discounted price process is a martingale under $\tilde{\mathbb{P}}$:

$$\tilde{\mathbb{E}}_n\left[\frac{V_{n+1}}{(1+r)^{n+1}}\right] = \frac{V_n}{(1+r)^n}$$

**$V_n$ is NOT Markov**

$V_n$ depends on both $S_n$ and $\max_{k \leq n} S_k$:

$$V_n = v_n\left(S_n, \max_{k \leq n} S_k\right)$$

Consider two paths that reach $S_2 = 4$ at time $n=2$:
- Path $HT$: $S_0=4 \to S_1=8 \to S_2=4$, $\max S = 8$
- Path $TH$: $S_0=4 \to S_1=2 \to S_2=4$, $\max S = 4$

Although $S_2$ is the same, $V_2(HT) \neq V_2(TH)$ because the ruuning maximum differs.
Hence the current state $S_n$ alone is not sufficient to determine $V_n$,
so $V_n$ is not Markov. $\blacksquare$

# Exercise 2.8 - Equivalence of Pricing Algorithms

### Problem
Show that the backward induction algorithm of Chatper 1(Theorem 1.2.2) gives the same derivative security prices as the risk-neutral pricing formula of Chapter 2 (equation 2.4.11).

### (i) Martingales are uniquely determined by theor terminal value

**Claim:** If $M_N = M'_N$ for all outcomes, then $M_n = M'_n$ for all $n$.

**Proof:** By the Martingale property $M_n = \tilde{\mathbb{E}}_n[M_N]$:

$$M_N = M'_n \implies \tilde{\mathbb{E}}_n[M_N] = \tilde{\mathbb{E}}_n[M'_N] \implies M_n = M'_n \quad \blacksquare$$

### (ii) Chapter 1 algorithm produces a Martingale

By algorithm (1.2.16):
$$V_n = \frac{1}{1+r}\tilde{\mathbb{E}}_n[V_{n+1}]$$

Dividing both sides by $(1+r)^n$:
$$\frac{V_n}{(1+r)^n} = \tilde{\mathbb{E}}_n\left[\frac{V_{n+1}}{(1+r)^{n+1}}\right] \quad \blacksquare$$

Hence $\frac{V_n}{(1+r)^n}$ is a martingale under $\tilde{\mathbb{P}}$.

### (iii) Risk-neutral pricing formula produces a Martingale

Define $V'_n = \tilde{\mathbb{E}}_n\left[\frac{V_N}{(1+r)^{N-n}}\right]$ by (2.4.11).

By the Tower Property:
$$\tilde{\mathbb{E}}_n\left[\frac{V'_{n+1}}{(1+r)^{n+1}}\right] = \tilde{\mathbb{E}}_n\left[\frac{1}{(1+r)^{n+1}}\tilde{\mathbb{E}}_{n+1}\left[\frac{V_N}{(1+r)^{N-n-1}}\right]\right]$$

$$= \tilde{\mathbb{E}}_n\left[\frac{V_N}{(1+r)^N}\right] = \frac{V'_n}{(1+r)^n} \quad \blacksquare$$

Hence $\frac{V'_n}{(1+r)^n}$ is a martingale under $\tilde{\mathbb{P}}$.

### (iv) The two algorithms are equivalent

- By (ii): $\frac{V_n}{(1+r)^n}$ is a martingale
- By (iii): $\frac{V'_n}{(1+r)^n}$ is a martingale
- At maturity: $V_N = V'_N$

By (i), two martingales with the same terminal value must be identival:

$$V_n = V'_n \quad \text{for all } n \quad \blacksquare$$

### Key Insight
This theorem bridges Chapter 1 and Chapter 2:

|Chapter 1|Chapter 2|
|---------|---------|
| Backward induction | Risk-neutral expectation |
| $V_n = \frac{1}{1+r}\tilde{\mathbb{E}}_n[V_{n+1}]$ | $V_n = \tilde{\mathbb{E}}_n\left[\frac{V_N}{(1+r)^{N-n}}\right]$ |
| Recursive (local) | Direct (global) |

Both give identical prices - the choice is a matter of computational convenience.

# Excercise 2.9 - Stochastic Volatility, Random Interst Rate

### Setup
From Figure 2.8.1:
- $S_0 = 4, r_0 = \frac{1}{4}$
- $S_1(H) = 8, r_1(H) = \frac{1}{4}$
- $S_1(T) = 2, r_1(T) = \frac{1}{2}$
- $S_2(HH) = 12, S_2(HT) = 8, S_2(TH) = 8, S_2(TT) = 2$

### (i) Risk-Neutral Probabilities

**At time 0:**
$$\tilde{p}_0 = \frac{1 + r_0 - d_0}{u_0 - d_0} = \frac{1 + \frac{1}{4} - \frac{1}{2}}{2 - \frac{1}{2}} = \frac{1}{2}, \quad \tilde{q}_0 = \frac{1}{2}$$

**At time 1 (H node):**
$$\tilde{p}_1(H) = \frac{1 + \frac{1}{4} - 1}{\frac{3}{2} - 1} = \frac{1}{2},  \quad \tilde{q}_1(H) = \frac{1}{2}$$

**At tume 1 (T node):**
$$\tilde{p}_1(T) = \frac{1 + \frac{1}{2} - 1}{4 - 1} = \frac{1}{6}, \quad \tilde{q}_1(T) = \frac{5}{6}$$

**Path probability:**

| Path | Probability |
|------|-------------|
| $HH$ | $\frac{1}{2} \cdot \frac{1}{2} = \frac{1}{4}$ |
| $HT$ | $\frac{1}{2} \cdot \frac{1}{2} = \frac{1}{4}$ |
| $TH$ | $\frac{1}{2} \cdot \frac{1}{6} = \frac{1}{12}$ |
| $TT$ | $\frac{1}{2} \cdot \frac{5}{6} = \frac{5}{12}$ |

### (ii) Option pricing with $V_2 = (S_2 - 7)^+$

**Payoffs at maturity:**
$$V_2(HH) = 5, \quad V_2(HT) = 1, \quad V_2(TH) = 1, \quad V_2(TT) = 0$$

**Backward induction:**

$$V_1(H) = \frac{1}{1+r_1(H)}[\tilde{p}_1(H) V_2(HH) + \tilde{q}_1(H) V_2(HT)] = \frac{4}{5}\left[\frac{1}{2} \cdot 5 + \frac{1}{2} \cdot 1 \right] = \frac{12}{5}$$

$$V_1(T) = \frac{1}{1+r_1(H)}[\tilde{p}_1(T) V_2(TH) + \tilde{q}_1(T9 V_2(TT)] = \frac{2}{3}\left[\frac{1}{6} \cdot 1 + \frac{5}{6} \cdot 0\right] = \frac{1}{9}$$

$$V_0 = \frac{1}{1+r_0}[\tilde{p}_0 V_1(H) + \tilde{q}_0 V_1(T)] = \frac{4}{5} \cdot \frac{1}{2}\left[\frac{12}{5} + \frac{1}{9}\right] = \frac{226}{225}$$

### (iii) Hedging at time 0

$$\Delta_0 = \frac{V_1(H) - V_1(T)}{S_1(H) - S_1(T)} = \frac{\frac{12}{5} - \frac{1}{9}}{8 - 2} = \frac{103}{270}$$

### (iv) Hedgin at time 1 (H node)

$$\Delta_1(H) = \frac{V_2(HH) - V_2(HT)}{S_2(HH) - S_2(HT)} = \frac{5 - 1}{12 - 8} = 1$$

# Exercise 2.10 - Dividend-Paying Stock

### Setup
Stock price evoloves as $S_{n+1} = (1 - A_{n+1})Y_{n+1}S_n$, where:
- $Y_{n+1} = u$ (head) or $d$ (tail)
- $A_{n+1} \in (0,1)$: dividend ratio
- Dividend paid at time $n+1$: $A_{n+1}Y_{n+1}S_n$

Wealth equation:
$$X_{n+1} = \Delta_n Y_{n+1} S_n + (1+r)(X_n - \Delta_n S_n)$$

### (i) Discounted Wealth Process is a Martingale

$$\tilde{\mathbb{E}}_n[X_{n+1}] = \Delta_n S_n [\tilde{p}u + \tilde{q}d] + (1 + r)(X_n - \Delta_n S_n)$$

Since $\tilde{p}u + \tilde{q}d = 1+r$:

$$ = (1+r)\Delta_n S_n + (1+r)(X_n - \Delta_n S_n) = (1+r)X_n$$

Therefore:
$$\frac{X_n}{(1+r}^n} = \tilde{\mathbb{E}}_n\left[\frac{X_{n+1}}{(1+r)^(n+1)\right] \quad \blacksquare$$

**Key Insight:** The dividend $\Delta_n A_{n+1}S_n$ is reinvested back into the portfolio, so it does not affect the wealth equation structure.

### (ii) Risk-Neutra; Pricing Formula Still Applies

By (i), $\frac{X_n}{(1+r)^n}$ is a martingale under $\tilde{\mathbb{P}}$.

The risk-neutral price process $\frac{V_n}{(1+r)^n}$ is also a martingale.

Since $X_N = V_N$ at maturity, by Exercise 2.8(i):

$$X_n = V_n \quad \text{for all } n \quad \blacksquare$$

### (iii) Discounted Stock Price is NOT a Martingale

$$\tilde{\mathbb{E}}_n\left[\frac{S_{n+1}}{(1+r)^{n+1}}\right] = \frac{S_n}{(1+r)^{n+1}}\tilde{\mathbb{E}}_n[(1-A_{n+1})Y_{n+1}]$$

In general, $\tilde{\mathbb{E}}_n[(1-A_{n+1} Y_{n+1}] \neq 1+r$, so:

$$\tilde{\mathbb{E}}_n\left[\frac{S_{n+1}}{(1+r)^{n+1}}\right] \neq \frac{S_n}{(1+r)^n} \quad \blacksquare$$

**Special case $A_{n+1} = a \in (0,1)$ (constant dividend rate):**

$$\tilde{\mathbb{E}}_n\left[\frac{S_{n+1}}{(1+r)^{n+1}}\right] = \frac{S_n}{(1+r)^{n+1}}(1-a)\tilde{\mathbb{E}}_n[Y_{n+1}] = \frac{S_n}{(1+r)^{n+1}}(1-a)(1+r) = (1-a)\frac{S_n}{(1+r)^n}$$

Multiplying both sides by $(1-a)^{-(n+1)}$:

$$\tilde{\mathbb{E}}_n\left[\frac{S_{n+1}}{(1-a)^{n+1}(1+r)^{n+1}}\right] = \frac{S_n}{(1-a)^n(1+r)^n} \quad \blacksquare$$

Hence $\frac{S_n}{(1-a)^n(1+r)^n}$ is a martingale under $\tilde{\mathbb{P}}$.

**Intuition:** The factor $\frac{1}{(1-a)^n}$ corrects for the cumulative dividend drain on the stock price, restoring the martingale property.

# Exercise 2.11 - Put-Call Prity

### Setup
In an $N$-period binomial model with no devidends:
- **Call:** $C_N = (S_N - K)^+$, $C_n = \tilde{\mathbb{E}}_n\left[\frac{C_N}{(1+r)^{N-n}}\right]$
- **Put:** $P_N = (K - S_N)^+$, $P_n = \tilde{\mathbb{E}}_n\left[\frac{P_N}{(1+r)^{N-n}}\right]$
- **Forward:** $F_N = S_N - K$, $F_n = \tilde{\mathbb{E}}_n\left[\frac{F_N}{(1+r)^{N-n}}\right]$

### (i) $C_N = F_N + P_N$

**Case $(S_N - K)^+ - (K - S_N)^+ = S_N - K$:

$$C_N - P_N = (S_N - K)^+ - (K - S_N)^+ = S_N - K = F_N$$

Hence $C_N = F_N + P_N$. $\blacksquare$

### (ii) $C_n = F_n + P_n$ for every $n$

By linearity of conditional expectations and (i):

$$C_n = \tilde{\mathbb{E}}_n\left[\frac{C_N}{(1+r)^{N-n}}\right] = \tilde{\mathbb{E}}_n\left[\frac{F_N + P_N}{(1+r)^{N-n}}\right] = F_n + P_n \quad \blacksquare$$

### (iii) $F_0 = S_0 - \frac{K}{(1+r)^N}$

By (ii): $F_0 = C_0 - P_o$. Using linearity and $(S_N-K)^+ - (K-S_N)^+ = S_N - K$:

$$F_0 = \tilde{\mathbb{E}}\left[\frac{S_N - K}{(1+r)^N}\right] = \tilde{\mathbb{E}}\left[\frac{S_N}{(1+r)^N}\right] - \frac{K}{(1+r)^N} = S_0 - \frac{K}{(1+r)^N} \quad \blacksquare$$

where the last equality uses the martingale property of the discounted stock price.

### (iv) Static Replication of the Forward Contract

At time zero, begin with $F_0$, buy one share, borrowing $\frac{K}{(1+r)^N}$:

$$X_0 = F_0 = S_0 - \frac{K}{(1+r)^N}$$

With no further trades, at time $N$:

$$X_N = S_n - (1+r)^N \cdot \frac{K}{(1+r)^N} = S_N - K = F_N \quad \blacksquare$$

**This is a static replication** - no rebalancing needed between time 0 and $N$.

### (v) Put-Call Parity at the forward Price

The forward price is $K = (1+r)^N S_0$, which gives $F_0 = 0$ by (iii):

$$F_0 = S_0 - \frac{(1+r)^N S_o}{(1+r)^N} = 0$$

By (ii):
$$C_0 = F_0 + P_0 = P_0 \quad \blacksquare$$

**At the forward strike price, call and put have the same price at time zero.**

### (vi) Does $C_n = P_n$ holde for every $n$?

By (ii), $C_n = P_n$ requires $F_n = 0$. Using the martingale property :

$$F_n = S_m - \frac{K}{(1+r)^{N-n}} = S_n - \frac{(1+r)^N S_0}{(1+r)^{N-n}} = S_n - (1+r)^n S_0$$

Since $S_n \neq (1+r)^n S_0$ in general for $n > 0$, we have $F_n \neq 0$, and therefore:

$$C_n \neq P_n \quad \text{for }n > 0 \text{ in general}\quad \blacksquare$$

**Intuition:** At time zero, $K = (1+r)^N S_0$ is the fair forward price, so call and put are symmetric. At late times, $S_n$ has moved away from $(1+r)^n S_0$, breaking the symmetry between call and put.

# Exercise 2.12 - Chooser Option

### Problem
A chooser option gives its owner the right to choose at time $m$  either:
- A call expiring at time $N$ with strike $K$
- A put expiring at time $N$ with strike $K$

Show that the time-zero price equals the sum of:
1. A put expiring at time $N$ with strike $K$
2. A call expring at time $m$ with strike $\frac{K}{(1+r)^{N-m}}$

### Solution

**Step 1: Payoff at time $m$**

The chooser option payoff at time $m$ is:

$$V_m = \max{C_m, P_m}$$

Using the identity $\max(a,b) = b + (a-b)^+$:

$$V_m = P_m + (C_m - P_m)^+$$

**step 2: Apply Put-Call Parity**

By Exercise 2.11(ii) and (iii):

$$V_m = P_m + \left(S_m - \frac{K}{(1+r)^{N-m}}\right)^+$$

The second term is the payoff of a call expiring at time $m$ with strike $\frac{K}{(1+r)^{N-m}}$.

**Step 3: Time-zero price**

$$V_0 = \tilde{\mathbb{E}}\left[\frac{V_m}{(1+r)^m}\right] 
= \tilde{\mathbb{E}}\left[\frac{P_m}{(1+r)^m}\right] 
+ \tilde{\mathbb{E}}\left[\frac{\left(S_m - \frac{K}{(1+r)^{N-m}}\right)^+}{(1+r)^m}\right]$$


**Step 4: Frist term equals $P_0$**

By the Tower Property:

$$\tilde{\mathbb{E}}\left[\frac{P_m}{(1+r)^m}\right] 
= \tilde{\mathbb{E}}\left[\frac{1}{(1+r)^m}
\tilde{\mathbb{E}}_m\left[\frac{(K-S_N)^+}{(1+r)^{N-m}}\right]\right]
= \tilde{\mathbb{E}}\left[\frac{(K-S_N)^+}{(1+r)^N}\right] = P_0$$


**Conclusion:**

$$V_0 = \underbrace{P_0}_{\text{put, expiry }N,\text{ strike }K} 
+ \underbrace{\tilde{\mathbb{E}}\left[\frac{\left(S_m - \frac{K}{(1+r)^{N-m}}\right)^+}{(1+r)^m}\right]}_{\text{call, expiry }m,\text{ strike }\frac{K}{(1+r)^{N-m}}} \quad \blacksquare$$

# Exercise 2.13 - Asian Option

### Setup
An Asian option has payoff:
$$V_N = f\left(\frac{1}{N+1}\sum_{n=0}^{N} S_n\right) = f\left(\frac{Y_N}{N+1}\right)$$

where $Y_n = \sum_{k=0}^{n} S_k$.

### (i) $(S_n, Y_n)$ is a Markov Process

We need to show \mathbb{E}_n[f(S_{n+1}, Y_{n+1})] = g(S_n, Y_n)$ for any function $f$.

Note that:
$$S_{n+1} = \begin{cases} uS_n & \omega_{n+1} = H \\ dS_n & \omega_{n+1} = T \end{cases}, \quad Y_{n+1} = Y_n + S_{n+1}$$

- $S_n, Y_n$ depend only on coin tosses $1, \ldots, n$ ->**known at time $n$**
- $\omega_{n+1}$ depends only on coin toss $n+1$ -> **independent of $\mathcal{F}_n$**

By Independence Lemma 2.5.3:

$$\mathbb{E}_n[f(S_{n+1}, Y_{n+1})] = \tilde{p}\, f(uS_n,\, Y_n + uS_n) + \tilde{q}\, f(dS_n,\, Y_n + dS_n) = g(S_n, Y_n) \quad \blacksquare$$

Since $\mathbb{E}_n[f(S_{n+1}, Y_{n+1})]$ depends only on $(S_n, Y_n)$,
the two-dimensional process $(S_n, Y_n)$ is Markov.

### (ii) Pricing Algorithm

**Terminal condition:**
$$v_n(s, y) = \frac{1}{1+r}\left[\tilde{p}\, v_{n+1}(us,\, y+us) + \tilde{q}\, v_{n+1}(ds,\, y+ds)\right]$$

**Connection to Exercise 1.8:**
For $f(x) = (x - K)^+$ (Asian call), $S_0=4, u=2, d=\frac{1}{2}[,r=\frac{1}{4}, N=3$:

$$v_3(s, y) = \left(\frac{y}{4} - K\right)^+$$

which is exactly the setup of Exercise 1.8, confirming the algorithm is consistent.

# Exercise 2.14 - Asian Option (Continued)

### Setup
An Asian Option with payoff based on the average of stock prices from time $M+1$ to $N$:

$$V_N = f\left(\frac{1}{N-M}\sum_{n=M+1}^{N} S_n\right) = f\left(\frac{Y_N}{N-M}\right)$$

where: 
$$Y_n = \begin{cases} 0 & 0 \leq n \leq M \\ \sum_{k=M+1}^{n} S_k & M+1 \leq n \leq N \end{cases}$$

### (i) $(S_n, Y_n)$ is a Markov Process

In both cases, $S_{n+1} = \omega_{n+1} S_n$ where $\omega_{n+1}$ is independent of $\mathcal{F}_n$.

**Case 1: $n < M$**

$Y_{n+1} = 0$ (deterministic), so by Independence Lemma 2.5.3:

$$\mathbb{E}_n[f(S_{n+1}, Y_{n+1})] = \tilde{p}\,f(uS_n, 0) +\tilde{q}\,f(dS_n, 0) = g(S_n, Y_n)$$

**Case 2: $n \geq M+1$**

$Y_{n+1} = Y_n + S_{n+1}$, so by Independence Lemma 2.5.3:

$$\mathbb{E}_n[f(S_{n+1}, Y_{n+1})] = \tilde{p}\, f(uS_n, Y_n+uS_n) + \tilde{q}\,f(dS_n, Y_n+dS_n) = g(S_n, Y_n)$$

In both cases, $\mathbb{E}_n[f(S_{n+1}, Y_{n+1})]$ depends only on $(S_n, Y_n)$,
so $(S_n, Y_n)$ is Markov. $blacksquare$

### (ii) Pricing Algorithm

**Terminal condition:**
$$v_N(s, y) = f\left(\frac{y}{N-M}\right)$$

**Case 1: $n \geq M+1$**
$$v_n(s, y) =\frac{1}{1+r}\left[\tilde{p}\, v_{n+1}(us,\ y+us) + \tilde{q}\, v_{n+1}(ds,\, y+ds)\right]$$

**Case 2: $n = M$ (transition)**

At time $M$, $Y_M = 0$ and $Y_{M+1} = S_{M+1}$:

$$v_M(s) =\frac{1}{1+r}\left[\tilde{p}\, v_{M+1}(us,\, us) + \tilde{q}\, v_{M+1}(ds,\, ds)\right]$$ 

**Case 3: $n < M$**

$Y_n = 0$ is deterministic, so $V_n$ depends only on $S_n$:

$$v_n(s) = \frac{1}{1+r}\left[\tilde{p}\, v_{n+1}(us) +\tilde{q}\, v_{n+1}(ds)\right]$$

### Key Difference from Exercise 2.13

| | Ex 2.13 | Ex 2.14 |
|---|---|---|
| Averaging period | $0$ to $N$ | $M+1$ to $N$ |
| State for $n \leq M$ | $(S_n, Y/n)$ | $S_n$ only |
| Transition formula | None | $v_M(s)$ in terms of $v_{m+1}(s, y)$ |

The averaging only starts at time $M+1$, so before time $M$ the option
price depends only on $S_n$, making the computation more efficient.
