# Chapter 3 — State Prices: Summary

## Key Concepts

### 1. State Price Density
The state price density is defined as:
$$\zeta_N(\omega) = \frac{Z(\omega)}{(1+r)^N} = \frac{\tilde{\mathbb{P}}(\omega)}{\mathbb{P}(\omega)(1+r)^N}$$

- **State price:** $\zeta_m p_m = \frac{\tilde{\mathbb{P}}(\omega_m)}{(1+r)^N}$ (Arrow-Debreu price)
- **State price density process:** $\zeta_n = \frac{Z_n}{(1+r)^n}$

### 2. Radon-Nikodym Derivative
$Z = \frac{d\tilde{\mathbb{P}}}{d\mathbb{P}}$ facilitates change of measure:
$$\tilde{\mathbb{E}}[Y] = \mathbb{E}[ZY], \quad \mathbb{E}[Y] = \tilde{\mathbb{E}}\left[\frac{Y}{Z}\right]$$

The Radon-Nikodym derivative process $Z_n = \mathbb{E}_n[Z]$ is a 
martingale under $\mathbb{P}$.

### 3. Risk-Neutral Pricing via State Prices
$$V_0 = \mathbb{E}[\zeta_N V_N] = \tilde{\mathbb{E}}\left[\frac{V_N}{(1+r)^N}\right]$$

More generally:
$$V_n = \frac{1}{\zeta_n}\mathbb{E}_n[\zeta_N V_N]$$

### 4. Optimal Investment Problem
Three equivalent formulations:

| Problem | Variable | Constraint |
|---------|----------|------------|
| 3.3.1 | $\Delta_0,\ldots,\Delta_{N-1}$ | Wealth equation |
| 3.3.3 | $X_N$ | $\tilde{\mathbb{E}}\left[\frac{X_N}{(1+r)^N}\right] = X_0$ |
| 3.3.5 | $(x_1,\ldots,x_M)$ | $\sum_m p_m \zeta_m x_m = X_0$ |

### 5. Solution via Lagrange Multipliers
First-order condition:
$$U'(X_N^*) = \frac{\lambda Z}{(1+r)^N} = \lambda\zeta_N$$

$$X_N^* = I\left(\frac{\lambda Z}{(1+r)^N}\right)$$

where $I = (U')^{-1}$ and $\lambda$ is determined by the budget constraint.

### 6. Optimal Wealth for Specific Utility Functions

| Utility $U(x)$ | Optimal $X_N^*$ |
|----------------|-----------------|
| $\ln x$ | $\frac{X_0(1+r)^N}{\lambda Z} = \frac{1}{\lambda\zeta_N}$ |
| $\frac{1}{p}x^p$ | $\frac{X_0(1+r)^N Z^{\frac{1}{p-1}}}{\mathbb{E}\left[Z^{\frac{p}{p-1}}\right]}$ |
| $\mathbf{1}_{x\geq\gamma}$ | $\gamma \cdot \mathbf{1}_{\zeta_N \leq \frac{1}{\lambda\gamma}}$ |

### 7. Optimality Verification (Exercise 3.8)
No Lagrange Multiplier Theorem needed. By concavity of $U$:
$$U(x) - yx \leq U(I(y)) - yI(y) \quad \forall x$$

Substituting $x = X_N$, $y = \frac{\lambda Z}{(1+r)^N}$ and taking expectations:
$$\mathbb{E}[U(X_N)] \leq \mathbb{E}[U(X_N^*)] \quad \blacksquare$$

## Three Equivalent Formulations of the Optimal Investment Problem
### Problem 3.3.1 - Portfolio Process Optimization

Given $X_0$, find an adapted portfolio process $\Delta_0, \Delta_1, \ldots, \Delta_{n-1}$
that maximizes:

$$\mathbb{E}[U(X_N)]$$

subject to the wealth equation:

$$X_{n+1} = \Delta_n S_{n+1} + (1+r)(X_n - \Delta_n S_n), \quad n = 0, 1, \ldots, N-1$$
#
### Problem 3.3.3 - Terminal Wealtj Optimization

Given $X_0$, find a random variable $X_n$ that maximizes:

$$\mathbb{E}[U(X_N)]$$

subject to the budget constraint:

$$\tilde{\mathbb{E}}\left[\frac{X_N}{(1+r)^N}\right] = X_0$$
#
### Problem 3.3.5 - Finite Sample Space Optimization

Gicen $X_0$, find a vector $(x_1, x_2, \ldots, x_M)$ that maximizes:

$$\sum_{m=1}^{M} p_m U(x_m)$$

subject to:

$$\sum_{m=1}^{M} p_m x_m \zeta_m = X_0$$

where $\Omega = \{\omega_1, \ldots, \omega_M\}$ is the finite sample space, $Z(\omega_m)$ is Radon Nikodym Derivatibe of $\tilde{\mathbb{P}}$ with respect to $\mathbb{P}$, 
$p_m = \mathbb{P}(\omega_m)$, and $\zeta_m$ is the state price density random variable:

$$\zeta_m = \frac{\tilde{\mathbb{P}}(\omega_m)}{\mathbb{P}(\omega_m)(1+r)^N} = \frac{Z(\omega_m)}{(1+r)^N}$$

## Equivalence of the Three Problems

### Proof: Problem 3.3.1 $\Leftrightarrow$ Problem 3.3.3

We show that the two problems have the same set of feasible solutions and
the same objective function, hence the same optimal solution.

**($\Rightarrow$)** Suppose $\Delta_0, \Delta_1, \ldots, \Delta_{n+1}$ is feasible 
for Problem 3.3.1. Define $X_N$ by iterating the wealth equation. Then:

$$\tilde{\mathbb{E}}\left[\frac{X_N}{(1+r)^N}\right] = X_0$$

since the discounted wealth process $\frac{X_n}{(1+r)^n}$ is a martingale
under $\tilde{\mathbb{P}}$ (Theorem 2.4.5). Hence $X_N$ is feasible for Problem 3.3.3.

**($\Leftarrow$)** Suppose $X_N$ is feasible for Problem 3.3.3, i.e.:

$$\tilde{\mathbb{E}}\left[\frac{X_N}{(1+r)^N}\right] = X_0$$

We may regard $X_N$ as a derivative security with time-zero price $X_0$.
By Theorem 1.2.2 (complete market), there exists a unique replicating
portfolio process $\Delta_0, \Delta_1, \ldots, \Delta_{N-1}$ starting with
initial wealth $X_0$ and achieving $X_N$ at time $N$.
Hence $\Delta_0, \Delta_1, \ldots, \Delta_{N-1}$ is feasible for Problem 3.3.1.

Since both problems have the same objective $\mathbb{E}[U(X_N)]$ and the feasible set (via the bijection obove), they hae the same optimal solution. $\blacksquare$
#
### Proof; Problem 3.3.3 $\Leftrightarrow$ Problem 3.3.5

**Objective function:** Writing out $\mathbb{E}[U(X_N)]$ on the finite sample space $\Omega = \{\omega1, \ldots, \omega_M\}$:

$$\mathbb{E}[U(X_N)] = \sum_{m=1}^{M} p_m U(x_m)$$

which is exactly the objective of problem 3.3.5.

**Budget constraint:** Writing out the constraint of Problem 3.3.3:

$$\tilde{\mathbb{E}}\left[\frac{X_N}{(1+r)^N}\right] = \sum_{m=1}^{M} \tilde{\mathbb{P}}(\omega_m) \frac{x_m}{(1+r)^N}$$

$$= \sum_{m=1}^{M} \mathbb{P}(\omega_m) \cdot \frac{\tilde{\mathbb{P}}(\omega_m)}{\mathbb{P}(\omega_m)(1+r)^N} \cdot x_m$$

$$= \sum_{m=1}^{M} p_m \zeta_m x_m = X_0$$

where $\zeta_m = \frac{\tilde{\mathbb{P}}(\omega_m)}{\mathbb{P}(\omega_m)(1+r)^N}$
is the state price density random variable, which is exactly the constraint  of Problem 3.3.5.

Since the objective functions and constraints are identical,
the two problems are equivalent. $\blacksquare$

## Solution via Problem 3,3,5 - Numerical Example

### Parameters 
$$S_0 = 4, \quad u = 2, \quad d = \frac{1}{2}, \quad r = \frac{1}{4}, \quad p = {2}{3}, \quad q = \frac{1}{3}, \quad X_0 = 4, \quad U(x) = \ln x$$

### Step 1: compute State Price Densities $\zeta_m$

Risk-neutral probabilities: $\tilde{p} = \tilde{q} = \frac{1}{2}$

$$Z(\omega) = \frac{\tilde{\mathbb{P}}(\omega)}{\mathbb{P}(\omega)}, \quad \zeta_m = \frac{Z(\omega)}{(1+r)^N} \quad \text{(state price density)}$$

$$\zeta_m p_m = \frac{\tilde{\mathbb{P}}(\omega_m)}{(1+r)^N} \quad \text{(state price)}$$

| Path $\omega$ | $\mathbb{P}(\omega)$ | $\tilde{\mathbb{P}}(\omega)$ | $Z(\omega)$ | $\zeta(\omega)$ | $\zeta_m p_m$ (state price)|
|---|---|---|---|---|---|
| $HH$ | $\frac{4}{9}$ | $\frac{1}{4}$ | $\frac{9}{16}$ | $\frac{9}{25}$ | $\frac{4}{25}$ |
| $HT$ | $\frac{2}{9}$ | $\frac{1}{4}$ | $\frac{9}{8}$ | $\frac{18}{25}$ | $\frac{4}{25}$ |
| $TH$ | $\frac{2}{9}$ | $\frac{1}{4}$ | $\frac{9}{8}$ | $\frac{18}{25}$ | $\frac{4}{25}$ |
| $TT$ | $\frac{1}{9}$ | $\frac{1}{4}$ | $\frac{9}{4}$ | $\frac{36}{25}$ | $\frac{4}{25}$ |

### Step 2: Solve via Lagrange Multipliers

For $U(x) = \ln x$, the first-order condition (3.3.24) gives:

$$U'(X_N^*) = \frac{1}{X_N^*} = \frac{\lambda Z}{(1+r)^N} \implies X_N^* = \frac{(1+r)^N}{\lambda Z} = \frac{25}{4\lambda Z}$$

### Step 3: Determine $\lambda$ from Budget Constraint

$$\sum_m p_m \zeta_m x_m = X_0 \implies \lambda = \frac{1}{4}$$

### Step 4: Optimal Termina Wealth

$$X_2^*(HH) = \frac{100}{9}, \quad X_2^*(HT) = \frac{50}{9}, \quad X_2^*(TH) = \frac{50}{9}, \quad X_2^*(TT) \frac{25}{9}$$

### Step 5: Verify Consistency with Problem 3.3.1

These values are consistent with the optimal portfolio process $\Delta_0 = \frac{5}{9}$, $\Delta_1(H) = \frac{25}{54}$, $\Delta_1(T) = \frac{25}{27}$
obtained by directly solving Problem 3.3.1, confirming the equivalence of the three formulations. $\blacksquare$

# Exercise 3.1 - Switch from $\tilde{\mathbb{E}}$ to $\mathbb{E}$ via $\frac{1}{Z}$

### Problem
Show that $\frac{1}{Z}$ facilitates the switch from $\tilde{\mathbb{E}}$ to $\mathbb{E}$:

**(i')** $\tilde{\mathbb{P}}\left(\frac{1}{Z} > 0\right) = 1$

**(ii')** $\tilde{\mathbb{E}}\left[\frac{1}{Z}\right] = 1$

**(iii')** For any random variable $Y$: $\mathbb{E}[Y] = \tilde{\mathbb{E}}\left[\frac{Y}{Z}\right]$

### Proof 

**(i')** By Theorem 3.1.1 $\mathbb{P}(Z > 0) = 1$, which implies $Z(\omega) > 0$
for all $\omega$ with positive probability. Hence $\frac{1}{Z(\omega)} > 0$
for all such $\omega$, and therefore $\tilde{\mathbb{P}}\left(\frac{1}{Z} > 0\right) = 1$. $\blacksquare$

**(ii')**
$$\tilde{\mathbb{E}}\left[\frac{1}{Z}\right] = \sum_{\omega \in \Omega} \frac{1}{Z(\omega)} \tilde{\mathbb{P}}(\omega) = \sum_{\omega \in \Omega} \frac{\mathbb{P}(\omega)}{\tilde{\mathbb{P}}(\omega)} = \sum_{\omega \in \Omega} \mathbb{P}(\omega) = 1 \quad \blacksquare$$

**(iii')***
$$\mathbb{E}[Y] = \sum_{\omega \in \Omega} Y(\omega) \mathbb{P}(\omega) = \sum_{\omega \in \Omega} Y(\omega) \frac{\mathbb{P}(\omega)}{\tilde{\mathbb{P}}(\omega)} \tilde{\mathbb{P}}(\omega = \tilde{\mathbb{E}}\left[\frac{Y}{Z}\right] \quad \blacksquare$$

### Key Insight
$Z$ and $\frac{1}{Z}$ are the Radon-Nikodym derivatives for seitching between $\mathbb{P}$ and $\tilde{\mathbb{P}}$:

$$\mathbb{E}[Y] = \tilde{\mathbb{E}}\left[\frac{Y}{Z}\right], \quad \tilde{\mathbb{E}}[Y] = \mathbb{E}[ZY]$$

This is the discrete-time analogue of the continuous-time change of measure(Girsanov's theorem), which will appear in Shreve II.

# Exercise 3.2 - Construction of Probability Measures via Radon-Nikodym Derivative

### Setup
Let $\mathbb{P}$ be a probability measure on finite $\Omega$, allwoing $\mathbb{P}(\omega) = 0$.
Let $Z$ be a random variable with $\mathbb{P}(Z \geq 0) = 1$ and $\mathbb{E}[Z] = 1$.
Define $\tilde{\mathbb{P}}(\omega) = Z(\omega)\mathbb{P}(\omega)$.

### (i) $\tilde{\mathbb{P}}$ is a probabliity measure

$$\tilde{\mathbb{P}}(\Omega) = \sum_{\omega \in \Omega} Z(\omega)\mathbb{P}(\omega) = \mathbb{E}[Z] = 1 \quad \blacksquare$$

### (ii) $\tilde{\mathbb{E}}[Y] = \mathbb{E}[ZY]$

$$\tilde{\mathbb{E}}[Y] = \sum_{\omega \in \Omega} Y(\omega)\tilde{\mathbb{P}}(\omega) = \sum_{\omega \in \Omega} Y(\omega) Z(\omega) \mathbb{P}(\omega) = \mathbb{E}[ZY] \quad \blacksquare$$

### (iii) $\mathbb{P}(A) = 0 \implies \tilde{\mathbb{P}}(A) = 0$

If $\mathbb{P}(A) = 0$, then $\mathbb{P}(\omega) = 0$ for all $\omega \in A$. Therefore:

$$\tilde{\mathbb{P}}(A) = \sum_{\omega \in A} Z(\omega)\mathbb{P}(\omega) = 0 \quad \blacksquare$$

### (iv) Assume $\mathbb{P}(Z > 0) = 1$. Then $\tilde{\mathbb{P}}(A) = 0 \implies \mathbb{P}(A) = 0$

$$\tilde{\mathbb{P}}(A) = \sum_{\omega \in A} Z(\omega)\mathbb{P}(\omega) = 0$$

Since $\mathbb{P}(Z > 0) = 1$, we have $Z(\omega) > 0$ for all $\omega$ with 
$\mathbb{P}(\omega) > 0$. Therefore $\mathbb{P}(\omega) = 0$ for all $\omega \in A$, 
hence $\mathbb{P}(A) = 0$. $\blacksquare$

From (iii) and (iv), $\mathbb{P}$ and $\tilde{\mathbb{P}}$ are **equivalent**
under the assumption $\mathbb{P}(Z > 0) = 1$.

### (v) Equivalent measures agree on probability-one events

$$\mathbb{P}(A) =1 \iff \mathbb{P}(A^c) = 0 \iff \tilde{\mathbb{P}}(A^c) = 0 \iff \tilde{\mathbb{P}}(A) = 1 \quad \blacksquare$$

where the middle equivalence usees the fact that $\mathbb{P}$ and $\tilde{\mathbb{P}}$
are equivalent.

### (vi) Example where $\mathbb{P}$ and $\tilde{\mathbb{P}}$ are not equivalent

Let $\Omega = \{H, T\}$ with:
$$\mathbb{P}(H) = \mathbb{P}(T) = \frac{1}{2}, \quad Z(H) = 2, \quad Z(T) = 0$$

Verify:
- $\mathbb{E}[Z] = \frac{1}{2} \cdot 2 + \frac{1}{2} \cdot 0 = 1$
- $\mathbb{P}(Z \geq 0) = 1$ but $\mathbb{P}(Z > 0) \neq 1$

Then $$\tilde{\mathbb{P}}(T) = Z(T)\mathbb{P}(T) = 0 \cdot \frac{1}{2} = 0$$

but $\mathbb{P}(T) = \frac{1}{2} \neq 0$.

Hence $\tilde{\mathbb{P}}(\{T\}) = 0$ but $\mathbb{P}(\{T\}) \neq 0$,
so $\mathbb{P}$ and $\tilde{\mathbb{P}}$ are **not equivalent**. $\blacksquare$

# Exercise 3.3 - Martingale Property of Condition Expectations 

### Setup 
From Figure 3.1.1: $S_0=4, u=2, d=\frac{1}{2}$, with actual probabilities $p=\frac{2}{3}, q=\frac{1}{3}$.

Define $M_n = \mathbb{E}_n[S_3]$ for $n = 0, 1, 2, 3$.

### Values of $M_n$

**$n = 3$:** $M_3 = S_3$

**$n = 2$:**
$$M_2(HH) = 24, \quad M_2(HT) = M_2(TH) = 6, \quad M_2(TT) = \frac{3}{2}$$

**$n=1$:**
$$M_1(H) = \frac{2}{3} \cdot 24 + \frac{1}{3}\ \cdot 6 = 18, \quad M_1(T) = \frac{2}{3} \cdot 6 + \frac{1}{3}\cdot \frac{3}{2} = \frac{9}{2}$$

**$n = 0$:**
$$M_0 = \frac{2}{3} \cdot 18 + \frac{1}{3} \cdot \frac{9}{2} = \frac{27}{2}$$

### Verification; $M_n is a Martingale

By the Tower Property of conditional expectations:

$$M_n = \mathbb{E}_n[S_3] = \mathbb{E}_n[\mathbb{E}_{n+1}[S_3]] = \mathbb{E}_n[M_{n+1}] \quad \blacksquare$$

Hence $M_0, M_1, M_2, M_3$ is a martingale under $\mathbb{P}$.

### Key Insight 
This illustrates Theorem 3.2.1: successive conditional expectations og any random variable form a martingale. The result holds for any probability measure - here we used the actual measure $\mathbb{P}$, not the risk-neutral probatility measure $\tilde{\mathbb{P}}$.

# Exercise 3.4 - State Price Densities and Asian Option Pricing 

### Setup
From Figure 3.1.1: $S_0=4, u=2, d=\frac{1}{2}, r=\frac{1}{4}$,
with $p=\frac{2}{3}, q=\frac{1}{3}$, $\tilde{p} = \tilde{q}=\frac{1}{2}$.

From Figure 3.2.1: Radon-Nikodym process $Z_n$.

### (i) State Price Densities $\zeta_3(\omega) = \frac{Z_3(\omega)}{(1+r)^3}$

$(1+r)^3 = \left(\frac{5}{4}\right)^3 = \frac{125}{64}$

| Path $\omega$ | $Z_3(\omega)$ | $\zeta_3(\omega)$ |
|---|---|---|
| $HHH$ | $\frac{27}{64}$ | $\frac{27}{125}$ |
| $HHT, HTH, THH$ | $\frac{27}{32}$ | $\frac{54}{125}$ |
| $HTT, THT, TTH$ | $\frac{27}{16}$ | $\frac{108}{125}$ |
| $TTT$ | $\frac{27}{8}$ | $\frac{216}{125}$ |

### (ii) Time-zero price of Asian Option via formula (3.1.10)

Asian option payoff; $V_3 = \left(\frac{Y_3}{4} - 4\right)^+$ where $Y_3 = \sum_{k=0}^{3} S_k$

| Path $\omega$ | $\mathbb{P}(\omega)$ | $Y_3$ | $V_3$ | $\zeta_3$ | $\mathbb{P}(\omega)\zeta_3 V_3$ |
|--------------|---------------------|-------|-------|-----------|-------------------------------|
| $HHH$ | $\frac{8}{27}$ | $60$ | $11$ | $\frac{27}{125}$ | $\frac{88}{125}$ |
| $HHT=HTH=THH$ | $\frac{4}{27}$ each | $36$ | $5$ | $\frac{54}{125}$ | $\frac{40}{125}$ each |
| $HTT=THT=TTH$ | $\frac{2}{27}$ each | $18$ | $\frac{1}{2}$ | $\frac{108}{125}$ | $\frac{4}{125}$ each |
| $TTT$ | $\frac{1}{27}$ | $7.5$ | $0$ | $\frac{216}{125}$ | $0$ |

$$V_0 = \mathbb{E}[\zeta_3 V_3] = \frac{88}{125} + 3 \cdot \frac{40}{125} + 3 \cdot \frac{4}{125} + 0 = \frac{152}{125}$$

This matches $v_0(4,4) = 1.216$ from Exercise 1.8. $\blacksquare$

### (iii) State Price Density $\zeta_2(HT) = \zeta_2(TH)$

$$\zeta_2(HT) = \zeta_2(TH) = \frac{Z_2(HT)}{(1+r)^2} = \frac{9/8}{25/16} = \frac{18}{25}$$

### (iv) Option Prices via Risk-Neutral Pricing Formula (3.2.6)

**$V_2(HT)$:** After path $HT$, next steps are $HTH$ or $HTT$:

$$\mathbb{E}_2[\zeta_3 V_3](HT) 
= \frac{2}{3} \cdot \frac{54}{125} \cdot 2 + \frac{1}{3} \cdot \frac{108}{125} \cdot \frac{1}{2}
= \frac{216}{375} + \frac{54}{375} = \frac{18}{25}$$

$$V_2(HT) = \frac{1}{\zeta_2(HT)} \cdot \frac{18}{25} = \frac{25}{18} \cdot \frac{18}{25} = 1$$

**$V_2(TH)$:** After path $TH$, next steps are $THH$ or $THT$:

$$\mathbb{E}_2[\zeta_3 V_3](TH) 
= \frac{2}{3} \cdot \frac{54}{125} \cdot \frac{1}{2} + \frac{1}{3} \cdot \frac{108}{125} \cdot 0
= \frac{18}{125}$$

$$V_2(TH) = \frac{1}{\zeta_2(TH)} \cdot \frac{18}{125} = \frac{25}{18} \cdot \frac{18}{125} = \frac{1}{5}$$

### Key Insight
$V_2(HT) = 1 \neq \frac{1}{5} = V_2(TH)$ even though $S_2(HT) = S_2(TH) = 4$.

This is because the Asian option is **path-dependent**: 
$Y_2(HT) = 4+8+4 = 16 \neq 10 = 4+2+4 = Y_2(TH)$, 
so the option price depends on the entire history, not just the current stock price.

# Exercise 3.5 — Stochastic Volatility, Random Interest Rate

### Setup
From Exercise 2.9: $S_0=4, r_0=\frac{1}{4}$, $S_1(H)=8, r_1(H)=\frac{1}{4}$, 
$S_1(T)=2, r_1(T)=\frac{1}{2}$, $S_2(HH)=12, S_2(HT)=S_2(TH)=8, S_2(TT)=2$

Risk-neutral probabilities (from Exercise 2.9):
$$\tilde{\mathbb{P}}(HH)=\frac{1}{4}, \quad \tilde{\mathbb{P}}(HT)=\frac{1}{4}, \quad 
\tilde{\mathbb{P}}(TH)=\frac{1}{12}, \quad \tilde{\mathbb{P}}(TT)=\frac{5}{12}$$

Actual probabilities:
$$\mathbb{P}(HH)=\frac{4}{9}, \quad \mathbb{P}(HT)=\frac{2}{9}, \quad 
\mathbb{P}(TH)=\frac{2}{9}, \quad \mathbb{P}(TT)=\frac{1}{9}$$

### (i) Radon-Nikodym Derivative $Z = \frac{d\tilde{\mathbb{P}}}{d\mathbb{P}}$

$$Z(\omega) = \frac{\tilde{\mathbb{P}}(\omega)}{\mathbb{P}(\omega)}$$

| Path $\omega$ | $\tilde{\mathbb{P}}(\omega)$ | $\mathbb{P}(\omega)$ | $Z(\omega)$ |
|--------------|------------------------------|---------------------|-------------|
| $HH$ | $\frac{1}{4}$ | $\frac{4}{9}$ | $\frac{9}{16}$ |
| $HT$ | $\frac{1}{4}$ | $\frac{2}{9}$ | $\frac{9}{8}$ |
| $TH$ | $\frac{1}{12}$ | $\frac{2}{9}$ | $\frac{3}{8}$ |
| $TT$ | $\frac{5}{12}$ | $\frac{1}{9}$ | $\frac{15}{4}$ |

### (ii) Radon-Nikodym Derivative Process $Z_0, Z_1, Z_2$

$Z_2 = Z$ (terminal value).

$$Z_1(H) = \frac{2}{3} \cdot \frac{9}{16} + \frac{1}{3} \cdot \frac{9}{8} = \frac{3}{4}$$

$$Z_1(T) = \frac{2}{3} \cdot \frac{3}{8} + \frac{1}{3} \cdot \frac{15}{4} = \frac{3}{2}$$

$$Z_0 = \mathbb{E}[Z] = \frac{4}{9} \cdot \frac{9}{16} + \frac{2}{9} \cdot \frac{9}{8} 
+ \frac{2}{9} \cdot \frac{3}{8} + \frac{1}{9} \cdot \frac{15}{4} = 1$$

### (iii) Option Pricing without Risk-Neutral Measure

For $V_2 = (S_2 - 7)^+$:
$$V_2(HH)=5, \quad V_2(HT)=1, \quad V_2(TH)=1, \quad V_2(TT)=0$$

**$V_1(H)$:**
$$\mathbb{E}_1[Z_2 V_2](H) = \frac{2}{3} \cdot \frac{9}{16} \cdot 5 + \frac{1}{3} \cdot \frac{9}{8} \cdot 1 = \frac{9}{4}$$

$$V_1(H) = \frac{1}{Z_1(H)(1+r_1(H))} \cdot \frac{9}{4} 
= \frac{1}{\frac{3}{4} \cdot \frac{5}{4}} \cdot \frac{9}{4} = \frac{12}{5}$$

**$V_1(T)$:**
$$\mathbb{E}_1[Z_2 V_2](T) = \frac{2}{3} \cdot \frac{3}{8} \cdot 1 + \frac{1}{3} \cdot \frac{15}{4} \cdot 0 = \frac{1}{4}$$

$$V_1(T) = \frac{1}{Z_1(T)(1+r_1(T))} \cdot \frac{1}{4} 
= \frac{1}{\frac{3}{2} \cdot \frac{3}{2}} \cdot \frac{1}{4} = \frac{1}{9}$$

**$V_0$:**
$$V_0 = \mathbb{E}\left[\frac{Z_2}{(1+r_0)(1+r_1)}V_2\right]$$

$$= \frac{4}{9} \cdot \frac{9/16}{25/16} \cdot 5 + \frac{2}{9} \cdot \frac{9/8}{25/16} \cdot 1 
+ \frac{2}{9} \cdot \frac{3/8}{15/8} \cdot 1 + 0$$

$$= \frac{4}{5} + \frac{4}{25} + \frac{2}{45} = \frac{226}{225}$$

### Key Insight
All results match Exercise 2.9(ii), confirming that the state price formula:

$$V_0 = \mathbb{E}\left[\frac{Z_2}{(1+r_0)(1+r_1)}V_2\right]$$

gives the same prices as the risk-neutral pricing formula, even in the 
stochastic volatility, random interest rate model.

# Exercise 3.6 — Optimal Wealth Process for Log Utility

### Problem
In an $N$-period binomial model with $U(x) = \ln x$, show that the optimal 
wealth process is:

$$X_n^* = \frac{1}{\lambda \zeta_n}, \quad n = 0, 1, \ldots, N$$

where $\zeta_n$ is the state price density process.

### Proof

**Step 1: Optimal terminal wealth**

By the first-order condition (3.3.24) with $U'(x) = \frac{1}{x}$:

$$\frac{1}{X_N^*} = \frac{\lambda Z_N}{(1+r)^N} = \lambda \zeta_N 
\implies X_N^* = \frac{1}{\lambda \zeta_N}$$

**Step 2: Optimal wealth at time $n$ via risk-neutral pricing**

$$X_n^* = \tilde{\mathbb{E}}_n\left[\frac{X_N^*}{(1+r)^{N-n}}\right] 
= \tilde{\mathbb{E}}_n\left[\frac{1}{\lambda \zeta_N (1+r)^{N-n}}\right]$$

Substituting $\zeta_N = \frac{Z_N}{(1+r)^N}$:

$$= \tilde{\mathbb{E}}_n\left[\frac{(1+r)^N}{\lambda Z_N (1+r)^{N-n}}\right] 
= \frac{(1+r)^n}{\lambda} \tilde{\mathbb{E}}_n\left[\frac{1}{Z_N}\right]$$

**Step 3: Apply Lemma 3.2.6**

By Lemma 3.2.6 with $Y = 1$:

$$\tilde{\mathbb{E}}_n\left[\frac{1}{Z_N}\right] = \frac{1}{Z_n}\mathbb{E}_n[1] = \frac{1}{Z_n}$$

**Step 4: Conclude**

$$X_n^* = \frac{(1+r)^n}{\lambda Z_n} = \frac{1}{\lambda \zeta_n} \quad \blacksquare$$

### Key Insight
The optimal wealth is **inversely proportional to the state price density** $\zeta_n$.

- In expensive states (high $\zeta_n$): hold less wealth
- In cheap states (low $\zeta_n$): hold more wealth

This is the continuous-time analogue of the result that will appear in Shreve II,
where the optimal wealth process under log utility satisfies 
$X_t^* \propto \frac{1}{\zeta_t}$.

# Exercise 3.7 — Optimal Wealth for Power Utility

### Problem
In an $N$-period binomial model with $U(x) = \frac{1}{p}x^p$, $p < 1, p \neq 0$,
show that the optimal terminal wealth is:

$$X_N^* = \frac{X_0 (1+r)^N Z^{\frac{1}{p-1}}}{\mathbb{E}\left[Z^{\frac{p}{p-1}}\right]}$$

### Proof

**Step 1: First-order condition**

$U'(x) = x^{p-1}$, so by (3.3.24):

$$(X_N^*)^{p-1} = \frac{\lambda Z}{(1+r)^N}$$

$$X_N^* = \left(\frac{\lambda Z}{(1+r)^N}\right)^{\frac{1}{p-1}} 
= \frac{\lambda^{\frac{1}{p-1}}}{(1+r)^{\frac{N}{p-1}}} Z^{\frac{1}{p-1}}$$

**Step 2: Determine $\lambda$ from budget constraint**

$$\tilde{\mathbb{E}}\left[\frac{X_N^*}{(1+r)^N}\right] = X_0$$

$$\frac{\lambda^{\frac{1}{p-1}}}{(1+r)^{\frac{N}{p-1}+N}} \tilde{\mathbb{E}}\left[Z^{\frac{1}{p-1}}\right] = X_0$$

Using $\tilde{\mathbb{E}}[Y] = \mathbb{E}[ZY]$:

$$\tilde{\mathbb{E}}\left[Z^{\frac{1}{p-1}}\right] = \mathbb{E}\left[Z^{1+\frac{1}{p-1}}\right] = \mathbb{E}\left[Z^{\frac{p}{p-1}}\right]$$

Solving for $\lambda^{\frac{1}{p-1}}$:

$$\lambda^{\frac{1}{p-1}} = \frac{X_0 (1+r)^{\frac{N}{p-1}+N}}{\mathbb{E}\left[Z^{\frac{p}{p-1}}\right]}$$

**Step 3: Substitute back**

$$X_N^* = \frac{\lambda^{\frac{1}{p-1}}}{(1+r)^{\frac{N}{p-1}}} Z^{\frac{1}{p-1}}
= \frac{X_0 (1+r)^{\frac{N}{p-1}+N}}{(1+r)^{\frac{N}{p-1}} \mathbb{E}\left[Z^{\frac{p}{p-1}}\right]} Z^{\frac{1}{p-1}}$$

$$= \frac{X_0 (1+r)^N Z^{\frac{1}{p-1}}}{\mathbb{E}\left[Z^{\frac{p}{p-1}}\right]} \quad \blacksquare$$

### Connection to Exercise 3.6 (Log Utility)

Taking $p \to 0$ recovers the log utility result:
$$X_N^* \propto Z^{\frac{1}{p-1}} \xrightarrow{p \to 0} Z^{-1} = \frac{1}{Z} \propto \frac{1}{\zeta_N}$$

consistent with Exercise 3.6. $\blacksquare$

# Exercise 3.8 — Verification of Optimality via Concavity

### Problem
Show directly (without relying on the Lagrange Multiplier Theorem) that 
$X_N^* = I\left(\frac{\lambda Z}{(1+r)^N}\right)$ maximizes $\mathbb{E}[U(X_N)]$ 
subject to the budget constraint (3.3.19).

### (i) Pointwise inequality from concavity of $U$

Fix $y > 0$. The function $f(x) = U(x) - yx$ is maximized at $x^* = I(y)$ 
since $f'(x) = U'(x) - y = 0 \implies x = I(y)$.

Since $U$ is concave, the tangent line at $x^* = I(y)$ lies above $U$:

$$U(x) \leq U(x^*) + U'(x^*)(x - x^*)$$

Since $U'(I(y)) = y$:

$$U(x) \leq U(I(y)) + y(x - I(y))$$

Rearranging:

$$U(x) - yx \leq U(I(y)) - yI(y) \tag{3.6.3} \quad \blacksquare$$

### (ii) $\mathbb{E}[U(X_N)] \leq \mathbb{E}[U(X_N^*)]$

Substitute $x = X_N$ and $y = \frac{\lambda Z}{(1+r)^N}$ into (3.6.3):

$$U(X_N) - \frac{\lambda Z}{(1+r)^N} X_N \leq U(X_N^*) - \frac{\lambda Z}{(1+r)^N} X_N^*$$

Taking expectations of both sides:

$$\mathbb{E}[U(X_N)] - \mathbb{E}\left[\frac{\lambda Z}{(1+r)^N} X_N\right] 
\leq \mathbb{E}[U(X_N^*)] - \mathbb{E}\left[\frac{\lambda Z}{(1+r)^N} X_N^*\right]$$

Using $\mathbb{E}[ZY] = \tilde{\mathbb{E}}[Y]$:

$$\mathbb{E}\left[\frac{\lambda Z}{(1+r)^N} X_N\right] 
= \lambda \tilde{\mathbb{E}}\left[\frac{X_N}{(1+r)^N}\right] = \lambda X_0$$

$$\mathbb{E}\left[\frac{\lambda Z}{(1+r)^N} X_N^*\right] 
= \lambda \tilde{\mathbb{E}}\left[\frac{X_N^*}{(1+r)^N}\right] = \lambda X_0$$

where both equalities use the budget constraint (3.3.19) and (3.3.26).

Therefore:

$$\mathbb{E}[U(X_N)] - \lambda X_0 \leq \mathbb{E}[U(X_N^*)] - \lambda X_0$$

$$\mathbb{E}[U(X_N)] \leq \mathbb{E}[U(X_N^*)] \quad \blacksquare$$

### Key Insight
This proof uses only:
1. **Concavity of $U$** → pointwise inequality (3.6.3)
2. **Budget constraint** → both sides have the same $\lambda X_0$ term

No Lagrange Multiplier Theorem is needed — the optimality follows directly 
from the concavity of the utility function.

# Exercise 3.9 — Maximizing Probability of Reaching a Goal

### Problem
Maximize $\mathbb{P}(X_N \geq \gamma)$ subject to:
$$\tilde{\mathbb{E}}\left[\frac{X_N}{(1+r)^N}\right] = X_0, \quad X_n \geq 0, \quad n = 1, 2, \ldots, N$$

### (i) $X_N \geq 0 \implies X_n \geq 0$ for all $n$

By the risk-neutral pricing formula:

$$X_n = \tilde{\mathbb{E}}_n\left[\frac{X_N}{(1+r)^{N-n}}\right] \geq 0 \quad \blacksquare$$

since $X_N \geq 0$ and $(1+r)^{N-n} > 0$.

### (ii) Pointwise inequality for $U(x) = \mathbf{1}_{x \geq \gamma}$

Define $I(y) = \gamma$ if $0 < y < \frac{1}{\gamma}$, and $I(y) = 0$ if $y > \frac{1}{\gamma}$.

We show $U(x) - yx \leq U(I(y)) - yI(y)$ for all $x \geq 0$.

**Case 1: $0 < y < \frac{1}{\gamma}$** → $I(y) = \gamma$, $U(I(y)) - yI(y) = 1 - y\gamma > 0$

- $x \geq \gamma$: $U(x) - yx = 1 - yx \leq 1 - y\gamma$
- $0 \leq x < \gamma$: $U(x) - yx = -yx \leq 0 < 1 - y\gamma$ 

**Case 2: $y > \frac{1}{\gamma}$** → $I(y) = 0$, $U(I(y)) - yI(y) = 0$

- $x \geq \gamma$: $U(x) - yx = 1 - yx \leq 1 - y\gamma < 0$ 
- $0 \leq x < \gamma$: $U(x) - yx = -yx \leq 0$ 

Hence $U(x) - yx \leq U(I(y)) - yI(y)$ for all $x \geq 0$. $\blacksquare$

### (iii) Optimal $X_N^* = I\left(\frac{\lambda Z}{(1+r)^N}\right)$

Substitute $x = X_N$ and $y = \frac{\lambda Z}{(1+r)^N}$ into (ii):

$$U(X_N) - \frac{\lambda Z}{(1+r)^N} X_N \leq U(X_N^*) - \frac{\lambda Z}{(1+r)^N} X_N^*$$

Taking expectations and using $\mathbb{E}[ZY] = \tilde{\mathbb{E}}[Y]$:

$$\mathbb{E}[U(X_N)] - \lambda\tilde{\mathbb{E}}\left[\frac{X_N}{(1+r)^N}\right] 
\leq \mathbb{E}[U(X_N^*)] - \lambda\tilde{\mathbb{E}}\left[\frac{X_N^*}{(1+r)^N}\right]$$

By budget constraint, both second terms equal $\lambda X_0$:

$$\mathbb{E}[U(X_N)] \leq \mathbb{E}[U(X_N^*)] \quad \blacksquare$$

### (iv) Existence of $\lambda$ $\Leftrightarrow$ Existence of $K$

Order the paths so that $\zeta_1 \leq \zeta_2 \leq \cdots \leq \zeta_M$.

Since $I(\lambda\zeta_m) = \gamma$ if $\zeta_m < \frac{1}{\lambda\gamma}$ and $0$ if 
$\zeta_m > \frac{1}{\lambda\gamma}$, there exists $K$ such that:

$$\zeta_K < \frac{1}{\lambda\gamma} < \zeta_{K+1}$$

i.e., $\zeta_K < \zeta_{K+1}$.

Substituting into the budget constraint (3.6.4):

$$\sum_{m=1}^{M} p_m \zeta_m I(\lambda\zeta_m) 
= \gamma\sum_{m=1}^{K} p_m \zeta_m + 0 = X_0$$

$$\sum_{m=1}^{K} p_m \zeta_m = \frac{X_0}{\gamma} \quad \blacksquare$$

### (v) Optimal Terminal Wealth

From (iv):

$$X_N^*(\omega^m) = I(\lambda\zeta_m) = \begin{cases} \gamma & m \leq K \\ 0 & m \geq K+1 \end{cases} \quad \blacksquare$$

### Key Insight
The optimal strategy invests only in the **cheapest states** 
(low state price density $\zeta_m$):

- **Cheap states** ($\zeta_m$ small, $m \leq K$): receive $\gamma$ — goal achieved
- **Expensive states** ($\zeta_m$ large, $m \geq K+1$): receive $0$ — goal not achieved

This is intuitive: to maximize the probability of reaching $\gamma$, 
invest in states where the "cost per unit of probability" is lowest.
